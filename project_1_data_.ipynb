# Step 2.3: handle missing value and check duplicates

```python
products['product_category_name'] = products['product_category_name'].fillna('Unknown')
products['product_category_name'].isnull().sum()

products['product_name_lenght'] = products['product_name_lenght'].fillna(products['product_name_lenght'].median())

products['product_description_lenght'] = products['product_description_lenght'].fillna(products['product_description_lenght'].median())

products['product_photos_qty'] = products['product_photos_qty'].fillna(products['product_photos_qty'].median())

products.isnull().sum()
```

```python
products['product_weight_g'] = products['product_weight_g'].fillna(products['product_weight_g'].median())

products['product_length_cm'] = products['product_length_cm'].fillna(products['product_length_cm'].median())

products['product_height_cm'] = products['product_height_cm'].fillna(products['product_height_cm'].median())

products['product_width_cm'] = products['product_width_cm'].fillna(products['product_width_cm'].median())

products.isnull().sum()
```

# Step 2.5: Convert Date Columns  (pd.to_datetime())

# Step 3: Merge the Datasets (Create Master Dataset)

# Step 3.1: Merge Orders and Customers

```python
master_df = pd.merge(
    orders,
    customers,
    on='customer_id',
    how='left'
)
master_df.head()
```

# Step 3.2: Merge Order Items

```python
master_df = pd.merge(
    master_df,
    order_items,
    on='order_id',
    how='left'
)
master_df.head()
```

# Step 3.3: Merge Products

```python
master_df = pd.merge(
    master_df,
    products,
    on='product_id',
    how='left'
)

print(master_df.shape)
#master_df.head()
```

# Step 3.4: Merge Sellers

```python
master_df = pd.merge(
    master_df,
    sellers,
    on='seller_id',
    how='left'
)

print(master_df.shape)
#master_df.head()
```

# Step 3.5: Merge Payments

```python
payments_summary = payments.groupby('order_id').agg({
    'payment_value': 'sum',
    'payment_installments': 'max',
    'payment_type': lambda x: ', '.join(x.unique())
}).reset_index()

payments_summary.columns = ['order_id', 'total_payment_value', 'max_payment_installments', 'payment_types']

#payments_summary.head()
```

```python
master_df = pd.merge(
    master_df,
    payments_summary,
    on='order_id',
    how='left'
)

print(master_df.shape)
#master_df.head()
```

# Step 3.6: Merge Reviews

```python
reviews_summary = reviews.groupby('order_id').agg({
    'review_score': 'mean'
}).reset_index()
```

```python
master_df = pd.merge(
    master_df,
    reviews_summary,
    on='order_id',
    how='left'
)

print(master_df.shape)
#master_df.head()
```

# Step 3.7: Merge Category Translation

```python
master_df = pd.merge(
    master_df,
    translation,
    on='product_category_name',
    how='left'
)

print(master_df.shape)
#master_df.head()
```

# Save the Master Dataset

```python
master_df.to_csv("master_dataset.csv", index=False)

print("Master Dataset Created Successfully!")
print(master_df.shape)
```

# Step 5.2 line chart (monthly trend)

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Professional theme
sns.set_theme(style="whitegrid")

# Create month column (if not already created)
master_df['order_month'] = master_df['order_purchase_timestamp'].dt.to_period('M').astype(str)

# Calculate monthly revenue
monthly_sales = (
    master_df.groupby('order_month')['price']
    .sum()
    .reset_index()
)

# Create figure
plt.figure(figsize=(14,7))

# Plot line chart
ax = sns.lineplot(
    data=monthly_sales,
    x='order_month',
    y='price',
    marker='o',
    linewidth=3,
    color='#1f77b4'
)

# Add value labels
for x, y in zip(monthly_sales['order_month'], monthly_sales['price']):
    plt.text(
        x,
        y,
        f"{y:,.0f}",
        ha='center',
        va='bottom',
        fontsize=9,
        fontweight='bold'
    )

# Title and labels
plt.title(
    "Monthly Sales Trend",
    fontsize=18,
    fontweight='bold',
    pad=20
)

plt.xlabel(
    "Month",
    fontsize=13,
    fontweight='bold'
)

plt.ylabel(
    "Total Revenue",
    fontsize=13,
    fontweight='bold'
)

# Rotate x-axis labels
plt.xticks(rotation=45)

# Remove top and right borders
sns.despine()

plt.tight_layout()
plt.show()
```

# Step 5.4 histogram (order values)

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Professional theme
sns.set_theme(style="whitegrid")

# Calculate total order value for each order
order_values = (
    master_df.groupby('order_id')['price']
    .sum()
    .reset_index()
)

# Create figure
plt.figure(figsize=(12,7))

# Histogram
sns.histplot(
    data=order_values,
    x='price',
    bins=30,
    kde=True,
    color='steelblue',
    edgecolor='black',
    alpha=0.8
)

# Title and labels
plt.title(
    "Distribution of Order Values",
    fontsize=18,
    fontweight='bold',
    pad=20
)

plt.xlabel(
    "Order Value",
    fontsize=13,
    fontweight='bold'
)

plt.ylabel(
    "Number of Orders",
    fontsize=13,
    fontweight='bold'
)

# Add average order value line
avg_order = order_values['price'].mean()

plt.axvline(
    avg_order,
    color='red',
    linestyle='--',
    linewidth=2,
    label=f'Average = {avg_order:.2f}'
)

plt.legend()

sns.despine()
plt.tight_layout()
plt.show()
```

# Step 5.5 pie chart (review scores)

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Professional style
sns.set_theme(style="whitegrid")

# Count review scores
review_counts = (
    master_df['review_score']
    .dropna()
    .astype(int)
    .value_counts()
    .sort_index()
)

# Attractive colors
colors = ['#FF6B6B', '#FFA94D', '#FFD43B', '#69DB7C', '#4D96FF']

# Highlight only 5-star reviews
explode = [0, 0, 0, 0, 0.08]

# Create figure
plt.figure(figsize=(9, 9))

plt.pie(
    review_counts,
    labels=[f"{i} Star" for i in review_counts.index],
    autopct='%1.1f%%',
    startangle=90,
    colors=colors,
    explode=explode,
    shadow=True,
    wedgeprops={
        'edgecolor': 'white',
        'linewidth': 2
    },
    textprops={
        'fontsize': 12,
        'fontweight': 'bold'
    }
)

plt.title(
    "Customer Review Score Distribution",
    fontsize=18,
    fontweight='bold',
    pad=20
)

plt.axis('equal')
plt.tight_layout()
plt.show()
```

# Step 5.6 heatmap (category vs month)

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Professional style
sns.set_theme(style="white")

# Create month column
master_df['order_month'] = pd.to_datetime(
    master_df['order_purchase_timestamp']
).dt.strftime('%b')

# Define month order
month_order = ['Jan','Feb','Mar','Apr','May','Jun',
               'Jul','Aug','Sep','Oct','Nov','Dec']

master_df['order_month'] = pd.Categorical(
    master_df['order_month'],
    categories=month_order,
    ordered=True
)

# Top 10 categories by revenue
top_categories = (
    master_df.groupby('product_category_name_english')['price']
    .sum()
    .sort_values(ascending=False)
    .head(10)
    .index
)

# Filter top categories
heatmap_df = master_df[
    master_df['product_category_name_english'].isin(top_categories)
]

# Create pivot table
pivot_table = heatmap_df.pivot_table(
    values='price',
    index='product_category_name_english',
    columns='order_month',
    aggfunc='sum',
    fill_value=0
)

# Plot heatmap
plt.figure(figsize=(14,8))

sns.heatmap(
    pivot_table,
    cmap='YlGnBu',
    annot=True,
    fmt='.0f',
    linewidths=0.5,
    linecolor='white',
    cbar_kws={'label': 'Revenue'}
)

plt.title(
    "Revenue by Product Category and Month",
    fontsize=18,
    fontweight='bold',
    pad=20
)

plt.xlabel(
    "Month",
    fontsize=13,
    fontweight='bold'
)

plt.ylabel(
    "Product Category",
    fontsize=13,
    fontweight='bold'
)

plt.tight_layout()
plt.show()
```

# Step 6: Business Insights Report

# 1. Revenue by Product Category

Finding (Bar Chart – Revenue by Product Category):
The Revenue by Product Category bar chart shows that a few product categories generate a significantly higher share of total revenue than the others, indicating that customer demand is concentrated in these categories.

Business Recommendation:
Increase inventory, marketing campaigns, and promotional activities for the top-performing categories while using discounts and bundle offers to improve sales of lower-performing categories.



# 2. Monthly Sales Trend

Finding (Line Chart – Monthly Sales Trend):
The Monthly Sales Trend chart shows noticeable fluctuations in revenue throughout the year, with one month generating the highest sales.

Business Recommendation:
Prepare additional inventory and launch marketing campaigns before peak sales months. During slower months, introduce seasonal discounts and special offers to maintain consistent sales.

# 3. Regional Sales Performance

Finding (Regional Sales Map/Bar Chart):
The Regional Sales visualization indicates that certain states contribute much more revenue than others, highlighting geographical differences in customer demand.

Business Recommendation:
Expand advertising and logistics support in high-performing regions while implementing targeted promotional campaigns in lower-performing regions to increase market penetration.

# 4. Customer Satisfaction

Finding (Pie Chart – Review Score Distribution):
The Review Score Distribution pie chart shows that most customers have given positive ratings, indicating overall customer satisfaction. However, a small percentage of low ratings still exists.

Business Recommendation:
Continue maintaining product quality and customer service while analyzing low-rated orders to identify issues related to delivery, product quality, or customer support and implement corrective actions.

# 5. Order Value Analysis

Finding (Histogram – Order Value Distribution & Heatmap – Category vs Month):
The Histogram shows that most orders fall within a moderate price range, while the Heatmap highlights seasonal demand patterns across product categories.

Business Recommendation:
Introduce product bundles, cross-selling, and upselling strategies to increase the average order value. Use the seasonal trends identified in the heatmap to optimize inventory planning and schedule promotions for high-demand categories during peak months.

# Most Surprising Finding

The most surprising finding was that most customers gave 4-star and 5-star reviews, indicating high customer satisfaction, but sales still varied significantly across different months and regions. This shows that positive customer reviews alone do not guarantee high sales. Other factors such as seasonal demand, marketing efforts, and regional preferences have a major impact on business performance. Therefore, the company should combine strong customer service with targeted marketing and better inventory planning to maximize revenue
